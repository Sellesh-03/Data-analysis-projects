```python
import pandas as pd

df = pd.read_csv(r"C:\Users\selle\Downloads\athlete_events.csv")
df
```

```python
df_copy = df.copy()
```

```python
df = df.drop_duplicates() 
df
```

```python
df['Name'] = df['Name'].str.replace(r'\s*\(.*?\)', '', regex=True)
df
```

```python
df['Sex'] = df['Sex'].replace({"M":"Male","F":"Female"})
df.head()
```

```python
df[df['Age'].isnull()]
```

```python
df = df.dropna().reset_index(drop=True)
df
```

```python
df[['Age','Height','Weight']] = df[['Age','Height','Weight']].astype(int)
df
```

```python
df.drop(columns = "Games")
```

```python
df[['First_Name','Last_Name']] = df['Name'].str.split(n=1,expand = True)
df
```

```python
new_order = ['ID','First_Name','Last_Name','Sex','Age','Height','Weight',
             'Team','NOC','Year','Season','City','Sport','Event','Medal'] 
df = df[new_order]
df
```

```python
df.drop(['Height','Weight'],axis=1,inplace = True)
df
```

```python
medals_by_country = df['NOC'].value_counts()
print(medals_by_country)
```

```python
medals_by_type = df.groupby(['NOC','Medal']).size().unstack(fill_value = 0)
print(medals_by_type)
```

```python
Gender_count = df['Sex'].value_counts()
print(Gender_count)
```

```python
import matplotlib.pyplot as plt

gender_count = df['Sex'].value_counts()
gender_count.plot(kind='pie', labels=['Male', 'Female'], autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution of Athletes')
plt.ylabel('')
plt.show()
```

```python
medals_by_gender = df.groupby(['Sex','Medal']).size().unstack(fill_value = 0)
print(medals_by_gender)
```

```python
game_count = df['Sport'].value_counts()
print(game_count)
```
